### Radians and degrees
* The radian measure $\theta$ of a counterclockwise rotation is the length along the unit circle of the arc traversed by P as it rotates into Q
* To convert from radians to degrees, multiply by $180 / \pi$
* To convert from degrees to radians, multiply by $\pi / 180$
| __Radians__ | __Degrees__ |
|-------------|-------------|
| 0           | 0&deg;      |
| $\frac{\pi}{6}$ | 30&deg; |
| $\frac{\pi}{4}$ | 45&deg; |
| $\frac{\pi}{3}$ | 60&deg; |
| $\frac{\pi}{2}$ | 90&deg; |
* TAn angle of $\theta$ subtends an arc of length $\theta r$ on a circle of radius $r$

### Sine and Cosine
* $sin\theta$ and $cos\theta$ can be defined in terms of right triangles:
  * $sin\theta = opposite/adjacent$
  * $cos\theta = adjacent/hypotenuse$
* This definition only makes since if $0 < \theta < \frac{\pi}{2}$, but they can be defined in terms of the unit circle.
* Let $P = (x,y)$ be the point on the unit circle corresponding to angle $\theta$:
  * $cos\theta = P_x$
  * $sin\theta = P_y$

* Sine and Cosine are __periodic__ functions with period $T=2\pi$
  * a function is periodic with period $T$ if $f(x+T)=f(x)$ for all $x$

* $sin(-\theta) = -sin\theta$
* $con(-\theta) = cos\theta$
 
### Trigonometric Identities
* $sin^2x + cos^2x = 1$ _* this is equivelant to the Pythagorean Theorem_
* Equivelant versions are obtained by dividing this equation by $cos^2x$ or $sin^2x$
  * $tan^2x+1 = sec^2x$
  *  $1+cot^2x=csc^2x$
* Complementary angles identities:
  * $sin(\frac{\pi}{2} - x) = cosx$
  * $cos(\frac{\pi}{2} - x) = sinx$
* Addition Forumulas:
  * $sin(x+y) = sinx *cosy + cosx*siny$
  * $cos(x+y) = cosx*cosy + sinx*siny$
* Double-angle Formulas:
  * $sin^2x=\frac{1}{2}(1-cos2x)$
  * $cos^2x=\frac{1}{2}(1+cos2x)$
  * $cos2x = cos^2x - sin^2x$
  * $sin2x = 2sinx*cosx$
* Shift Formulas:
  * $sin(x+\frac{\pi}{2}) = cosx$
  * $cos(x+\frac{\pi}{2}) = -sinx$

### Law of Cosines
If a triangle has sides $a$, $b$, and $c$ and $\theta$ is the angle opposite cide $c$, then:
* $c^2=a^2+b^2-2ab*cos\theta$

This is a generalization of the Pythagorean theorm.
* If $\theta=90^\circ$ then $cos\theta=0$ and the Law of Cosines reduces to the pythagorean theorem

### Four additional trigonometric functions:
* $tan\theta = \frac{sin\theta}{cos\theta}$
* $cot\theta = \frac{cos\theta}{sin\theta}$
* $sec\theta = \frac{1}{cos\theta}$
* $csc\theta = \frac{1}{sin\theta}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate x values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 1000)
y = np.sin(x)  # Compute sine values

# Plot the sine wave
plt.plot(x, y, label='sin(x)', color='b')
plt.xlabel("x (radians)")
plt.ylabel("sin(x)")
plt.title("Sine Wave")
plt.axhline(0, color='black', linewidth=0.5)  # Add horizontal axis
plt.axvline(0, color='black', linewidth=0.5)  # Add vertical axis
plt.grid(True, linestyle='--', alpha=0.6)  # Add grid
plt.legend()
```

```python
# Generate x values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 1000)
y = np.cos(x)  # Compute sine values

# Plot the sine wave
plt.plot(x, y, label='sin(x)', color='b')
plt.xlabel("x (radians)")
plt.ylabel("cos(x)")
plt.title("Cosine Wave")
plt.axhline(0, color='black', linewidth=0.5)  # Add horizontal axis
plt.axvline(0, color='black', linewidth=0.5)  # Add vertical axis
plt.grid(True, linestyle='--', alpha=0.6)  # Add grid
plt.legend()
```
