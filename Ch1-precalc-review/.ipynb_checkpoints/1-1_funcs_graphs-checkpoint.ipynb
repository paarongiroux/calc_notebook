# Chapter 1.1 - Real Numbers, Functions and Graphs
### Triangle Inequalities
* $|a + b| \leq |a| + |b|$

### Open and Closed Intervals
* $(a,b) = {x : |x - c| < r}$
* $[a,b] = {x : |x - c| \leq r}$

where __$ c = \frac{1}{2} (a+b)$__ is the midpoint and __$r = \frac{1}{2} (b-a)$__ is the radius

### Distance between two points
d = $ \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2} $

### Equation of a circle of radias $ r $ and center $ (a,b) $
$ (x-a)^2 + (y-b)^2 = r^2 $

### Function parity
* Even function: $ f(-x) = f(x) $ _\*(graph is symmetric about the y-axis)_
* Odd function: $(f-x) = -f(x)$ _\*(graph is symmetric about the origin)_

### Translation and Scaling
* __Vertical Translation__ $y = f(x) + c$: shifts the graph by $|c|$ units _vertically_
* __Horizontal Translation__ $y = f(x+c)$: shifts the graph by $|c|$ units _horizontally_


* __Vertical Scaling__ $y = kf(x)$
  * $k > 1$: graph is expanded vertically by a factor of $k$
  * $0 < k < 1$: graph is compressed vertically by factor of $k$
  * $k<0$: graph is also reflected across x axis


* __Horizontal Scaling__ $y = f(kx)$
  * $k > 1$: graph is compressed horizontally by a factor of $k$
  * $0 < k < 1$: graph is expanded horizontally by a factor of $k$
  * $k < 0$: graph is also reflected across y axis

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function
def f(x):
    return x

# Generate x values
x = np.linspace(-10, 10, 100)  # 100 points between -10 and 10

# Calculate y values
y = f(x)

# Plot the function
plt.plot(x, y, label='f(x) = x')

# Customize the graph
plt.title("Odd function: f(x) = x")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.8)  # x-axis
plt.axvline(0, color='black', linewidth=0.8)  # y-axis
plt.grid(alpha=0.4)
plt.legend()

# Show the plot
plt.show()
```

```python
def f(x):
    return x * x

# Generate x values
x = np.linspace(-10, 10, 100)  # 100 points between -10 and 10

# Calculate y values
y = f(x)

# Plot the function
plt.plot(x, y, label='f(x) = x^2')

# Customize the graph
plt.title("Even function: f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.8)  # x-axis
plt.axvline(0, color='black', linewidth=0.8)  # y-axis
plt.grid(alpha=0.4)
plt.legend()

# Show the plot
plt.show()
```

```python
c = 50
def f(x):
    return (x * x) + c

# Generate x values
x = np.linspace(-10, 10, 100)  # 100 points between -10 and 10

# Calculate y values
y = f(x)

# Plot the function
plt.plot(x, y, label='f(x) = x^2 + 50')

# Customize the graph
plt.title("Vertical Translation (+50)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.8)  # x-axis
plt.axvline(0, color='black', linewidth=0.8)  # y-axis
plt.grid(alpha=0.4)
plt.legend()

# Show the plot
plt.show()
```

```python
c = 1

def f(x):
    return (x + c) * (x + c)

# Generate x values
x = np.linspace(-10, 10, 100)  # 100 points between -10 and 10

# Calculate y values
y = f(x)

# Plot the function
plt.plot(x, y, label='f(x) = (x-1)^2')

# Customize the graph
plt.title("Tanslate left by 1")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.8)  # x-axis
plt.axvline(0, color='black', linewidth=0.8)  # y-axis
plt.grid(alpha=0.4)
plt.legend()

# Show the plot
plt.show()
```

```python
k = 3
def f(x):
    return k * np.sin(x)

# Generate x values
x = np.linspace(-10, 10, 100)  # 100 points between -10 and 10

# Calculate y values
y = f(x)

# Plot the function
plt.plot(x, y, label='f(x) = k * sin(x)')

# Customize the graph
plt.title("Vertical scaling")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.8)  # x-axis
plt.axvline(0, color='black', linewidth=0.8)  # y-axis
plt.grid(alpha=0.4)
plt.legend()

# Show the plot
plt.show()
```

```python
k = 3
def f(x):
    return np.sin(k * x)

# Generate x values
x = np.linspace(-10, 10, 100)  # 100 points between -10 and 10

# Calculate y values
y = f(x)

# Plot the function
plt.plot(x, y, label='f(x) = k * sin(x)')

# Customize the graph
plt.title("Horizontal scaling")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.8)  # x-axis
plt.axvline(0, color='black', linewidth=0.8)  # y-axis
plt.grid(alpha=0.4)
plt.legend()

# Show the plot
plt.show()
```
