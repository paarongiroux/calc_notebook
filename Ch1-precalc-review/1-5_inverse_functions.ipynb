# Chapter 1.5 Inverse Functions
The inverse of $f(x)$, denoted $f^{-1}(x)$ is the function that _reverses_ the effect of $f(x)$

### Inverse Function Definition
Let $f(x)$ have domain $D$ and range $R$. If there is a function $g(x)$ with domain $R$ such that
* $g(f(x))=x$ for $x \in D$ and $f(g(x)) = x$ for $x \in R$

then $f(x)$ is said to be __invertible__. The function $g(x)$ is called the __inverse functon__

### One-to-One Functions
* A function $f(x)$ is one-to-one on domain $D$ if, for every value $c$, the equation $f(x) = x$ has at most one solution for $x \in D$.
* In otherwords $\forall a,b \in D: f(a)=f(b) \iff a=b$
* When $f(x)$ is one-to-one on its domain $D$, the inverse exists and its domain is equal to the range $R$ of $f$

### Graphing an inverse function
* The __reflection__ of point $(a,b)$ through the line $y=x$ is the point $(b,a)$
* The graph of $f^{-1}$ is the reflection of the graph of $f$ through $y=x$

### Inverse Tigonometric Functions
* The trigonometrics functions are not one-to-one, therefore, we must restrict the domains in order to find an inverse.
* First, consider the sine function. $f(\theta) = sin\theta$ is one-to-one on $[-\frac{\pi}{2}, \frac{\pi}{2}]$
  * With this interval as the domain, the inverse is called the __arcsine function__ and is denoted $\theta = sin^{-1}x$ or $\theta = arcsin(x)$
  * The __arccosine function__ can be defined similarly
* $\theta = sin^{-1}x$ is the unique angle in $[-\frac{\pi}{2}, \frac{\pi}{2}]$ such that $sin\theta = x$
* $\theta = cos^{-1}x$ is the unique angle in $[0, \pi]$ such that $cos\theta = x$
* $\theta = tan^{-1}x$ is the unique angle in $[-\frac{\pi}{2}, \frac{\pi}{2}]$ such that $tan\theta = x$
* $\theta = cot^{-1}x$ is the unique angle in $[0, \pi]$ such that $cot\theta = x$
* $\theta = sec^{-1}x$ is the unique angle in $[0, \frac{\pi}{2}) \cup (\frac{\pi}{2},\pi]$ such that $sec\theta = x$
* $\theta = csc^{-1}x$ is the unique angle in $[-\frac{\pi}{2},0) \cup (0, \frac{\pi}{2}]$ such that $csc\theta = x$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x * x * x

def inverseF(x):
    return np.cbrt(x)
```

```python
x = np.linspace(-10, 10, 100)  

y = f(x)

# Plot the function
plt.plot(x, y, label='f(x) = x^3')

# Customize the graph
plt.title("f(x) = x^3")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.8)  # x-axis
plt.axvline(0, color='black', linewidth=0.8)  # y-axis
plt.grid(alpha=0.4)
plt.legend()

# Show the plot
plt.show()
```

```python
x = np.linspace(-10, 10, 100)  

y = inverseF(x)

# Plot the function
plt.plot(x, y, label='f^-1(x) = cbrt(3)')

# Customize the graph
plt.title("f^-1(x) = cubrt(3)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.8)  # x-axis
plt.axvline(0, color='black', linewidth=0.8)  # y-axis
plt.grid(alpha=0.4)
plt.legend()

# Show the plot
plt.show()
```
